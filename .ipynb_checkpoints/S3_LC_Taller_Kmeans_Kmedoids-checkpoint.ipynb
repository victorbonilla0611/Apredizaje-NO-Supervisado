## Desarrollo


### 1.Carga de datos 

Los datos pueden ser bajados directamente utilizando la librería `scikit-learn`,  la función que le permite hacer esto es `fetch_olivetti_faces`. 

```python
from sklearn.datasets import fetch_olivetti_faces

# Descargamos el dataset Olivetti Faces
faces = fetch_olivetti_faces(shuffle=True, random_state=312)

# Variables
X = faces.data      
y = faces.target    
images = faces.images  

print("Shape de X:", X.shape)
print("Shape de y:", y.shape)
print("Número de imágenes:", images.shape[0])
```

Se inicio utilizando la función fetch_olivetti_faces de la librería scikit-learn, que permite descargar directamente la base de datos Olivetti Faces. Esta contiene 400 imágenes en escala de grises correspondientes a 40 individuos. 

### 2.  Análisis preliminar 

#### 2.1. Describa los contenidos de la base de datos, tenga en cuenta que debe detallar los elementos, la dimensión de la base, y la dimensión de las imágenes. 

```python
# Análisis preliminar
import matplotlib.pyplot as plt

print("Análisis preliminar del dataset Olivetti Faces")
print("--------------------------------------------------")
print(f"➡ Número total de imágenes: {X.shape[0]}")
print(f"➡ Número de individuos: {len(set(y))}")
print(f"➡ Imágenes por persona: {X.shape[0] // len(set(y))}")
print(f"➡ Dimensión de cada imagen: {images.shape[1]}x{images.shape[2]} píxeles")
print(f"➡ Cada imagen equivale a un vector de: {X.shape[1]} características")
print("--------------------------------------------------")

# Mostrar 10 imágenes de ejemplo
fig, axes = plt.subplots(1, 10, figsize=(15, 4))
for i, ax in enumerate(axes):
    ax.imshow(images[i], cmap="gray")
    ax.axis("off")
    ax.set_title(y[i])
plt.show()
```

#### 2.2. Genere una figura con una imágen de cada uno de los individuos

```python
# Figura con una imagen de cada individuo
import numpy as np

print("Figura con una imagen de cada individuo:")
print("--------------------------------------------------")

n_individuos = len(set(y))  # Total de personas (40 en Olivetti Faces)
fig, axes = plt.subplots(4, 10, figsize=(15, 6))  # 4 filas x 10 columnas

for i in range(n_individuos):
    ax = axes[i // 10, i % 10]
    # Seleccionamos la primera imagen del individuo i
    idx = np.where(y == i)[0][0]
    ax.imshow(images[idx], cmap="gray")
    ax.axis("off")
    ax.set_title(f"ID {i}", fontsize=8)

plt.suptitle("Una imagen de cada individuo en Olivetti Faces", fontsize=14)
plt.show()
```

#### 2.3. Transforme las imágenes en un data frame donde cada columna contiene los valores estandarizados del nivel de gris en la imagen y una columna con la etiqueta de cada imagen. 

```python
# Transformación a DataFrame con estandarización
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Estandarizar los valores de gris de las imágenes
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Crear un DataFrame con las imágenes estandarizadas
df_faces = pd.DataFrame(X_scaled)

# Agregar la columna de la etiqueta (persona)
df_faces["label"] = y

# Información general
print("** Información del DataFrame de imágenes")
print(f"- Shape: {df_faces.shape[0]} filas, {df_faces.shape[1]} columnas")
print(f"- Número de individuos: {df_faces['label'].nunique()}")

# Mostrar primeras 5 filas con pocas columnas para legibilidad
print("\n** Vista de las primeras 5 filas (primeros 10 pixeles + etiqueta):")
display(df_faces.iloc[:, :10].join(df_faces["label"]).head())

# Estadísticas básicas de los pixeles
print("\n** Resumen estadístico de las intensidades estandarizadas (primeros 10 pixeles):")
display(df_faces.iloc[:, :10].describe().T.round(2))
```

### 3. Clasificando de forma no supervisada

En esta sección trataremos de clasificar las imágenes de los individuos de forma no supervisada. Para ello utilizaremos los algoritmos de k-medias y k-medoides. Dado que sabemos que el dataset consta de 40 personas diferentes, utilizaremos esta información para pedirle a los algoritmos que encuentren k=40 clusters y examinaremos su precisión.


#### 3.1. Implemente k-medias sobre los datos estandarizados en el punto anterior. Al implementar utilice  `random_state=123` y `n_init=10`. (Esto puede tomar mucho tiempo y requerir mucho RAM, puede aprovechar los recursos de [Google Colab](https://colab.research.google.com/))

```python
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score

# Número de clusters (Usaremos 40 como prueba porque sabemos que son 40 personas)
k = 40  

# Implementamos K-Means
kmeans = KMeans(n_clusters=k, random_state=123, n_init=10)
labels_kmeans = kmeans.fit_predict(df_faces.drop("label", axis=1))

# Evaluamos la calidad de los clusters
ari_score = adjusted_rand_score(df_faces["label"], labels_kmeans)

print("Número de clusters encontrados:", len(set(labels_kmeans)))
print("ARI (Adjusted Rand Index):", ari_score)
```

(Utilice este espacio para describir el procedimiento, análisis y conclusiones)

#### 3.2. Evalúe la precisión del algoritmo para agrupar las imágenes de los individuos. 

```python
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score

# etiquetas reales (ground truth)
true_labels = faces.target  

# etiquetas predichas por K-means
pred_labels = labels_kmeans  

# ARI
ari = adjusted_rand_score(true_labels, pred_labels)

# NMI
nmi = normalized_mutual_info_score(true_labels, pred_labels)

print(f"Adjusted Rand Index (ARI): {ari:.4f}")
print(f"Normalized Mutual Information (NMI): {nmi:.4f}")
```

#### 3.3. Usando PCA reduzca la dimensión de la matriz de predictores. Retenga el numero de componentes que explican el 95% de la varianza y vuelva a utilizar k-medias para clasificar las imágenes comentado si la precisión mejoró. 

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.95, svd_solver='full') 
X_pca = pca.fit_transform(X)
print("Componentes retenidos:", pca.n_components_)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score, accuracy_score
from scipy.stats import mode
import numpy as np

# Entrenar K-means
kmeans_pca = KMeans(n_clusters=40, random_state=123, n_init=10)
labels_pca = kmeans_pca.fit_predict(X_pca)

# Reetiquetar clusters para compararlos con las etiquetas reales
def cluster_accuracy(true_labels, cluster_labels):
    labels = np.zeros_like(cluster_labels)
    for i in range(40):  # 40 personas
        mask = (cluster_labels == i)
        if np.sum(mask) > 0:
            labels[mask] = mode(true_labels[mask], keepdims=False).mode
    return accuracy_score(true_labels, labels)

y_true = faces.target

acc_pca = cluster_accuracy(y_true, labels_pca)
ari_pca = adjusted_rand_score(y_true, labels_pca)
nmi_pca = normalized_mutual_info_score(y_true, labels_pca)

print("Precisión (PCA + k-means):", acc_pca)
print("ARI:", ari_pca)
print("NMI:", nmi_pca)
```

```python
from sklearn_extra.cluster import KMedoids

# Entrenar K-Medoides
kmedoids_pca = KMedoids(n_clusters=40, random_state=123, method="pam", metric="euclidean")
labels_kmedoids = kmedoids_pca.fit_predict(X_pca)

# Reetiquetar clusters para compararlos con las etiquetas reales
def cluster_accuracy(true_labels, cluster_labels):
    labels = np.zeros_like(cluster_labels)
    for i in range(40):  # 40 personas
        mask = (cluster_labels == i)
        if np.sum(mask) > 0:
            labels[mask] = mode(true_labels[mask], keepdims=False).mode
    return accuracy_score(true_labels, labels)

acc_kmedoids = cluster_accuracy(y_true, labels_kmedoids)
ari_kmedoids = adjusted_rand_score(y_true, labels_kmedoids)
nmi_kmedoids = normalized_mutual_info_score(y_true, labels_kmedoids)

print("Precisión (PCA + K-medoides):", acc_kmedoids)
print("ARI:", ari_kmedoids)
print("NMI:", nmi_kmedoids)
```

### 4. Número de clusters óptimo

En el punto anteriores utilizamos nuestro conocimiento previo sobre los datos para elegir el número de clusters. En este punto, verifique si 40 es realmente el número de cluster que usted elegiría según los criterios estudiados en los cuadernos teóricos. Discuta los resultados obtenidos.

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from sklearn_extra.cluster import KMedoids
import matplotlib.pyplot as plt

# Rango de clusters a evaluar
range_n_clusters = [10, 20, 30, 40, 50, 60]

sil_scores = []
ch_scores = []
db_scores = []

for k in range_n_clusters:
    kmedoids = KMedoids(n_clusters=k, random_state=123, method="pam", metric="euclidean")
    labels = kmedoids.fit_predict(X_pca)
    
    sil_scores.append(silhouette_score(X_pca, labels))
    ch_scores.append(calinski_harabasz_score(X_pca, labels))
    db_scores.append(davies_bouldin_score(X_pca, labels))

# Graficar resultados
plt.figure(figsize=(14,4))

plt.subplot(1,3,1)
plt.plot(range_n_clusters, sil_scores, marker='o')
plt.title("Coeficiente de Silueta")
plt.xlabel("Número de clusters")
plt.ylabel("Silueta")

plt.subplot(1,3,2)
plt.plot(range_n_clusters, ch_scores, marker='o')
plt.title("Índice Calinski-Harabasz")
plt.xlabel("Número de clusters")
plt.ylabel("CH Score")

plt.subplot(1,3,3)
plt.plot(range_n_clusters, db_scores, marker='o')
plt.title("Índice Davies-Bouldin")
plt.xlabel("Número de clusters")
plt.ylabel("DB Score (menor es mejor)")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
from sklearn_extra.cluster import KMedoids

# Rango de posibles k
k_values = range(2, 60)  # puedes ajustar el rango
wcss = []   # Suma de distancias intra-cluster
silhouette_scores = []

for k in k_values:
    kmedoids = KMedoids(n_clusters=k, random_state=123, method="pam", metric="euclidean")
    labels = kmedoids.fit_predict(X_pca)
    
    # WCSS (suma de distancias al medoid de cada cluster)
    inertia = np.sum(np.min(kmedoids.transform(X_pca), axis=1))
    wcss.append(inertia)
    
    # Silhouette solo si k>1
    if k > 1:
        sil_score = silhouette_score(X_pca, labels)
        silhouette_scores.append(sil_score)
    else:
        silhouette_scores.append(np.nan)

# --- Graficar resultados ---
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(k_values, wcss, marker='o')
plt.title("Suma de varianzas intra-cluster (WCSS)")
plt.xlabel("Número de clusters (k)")
plt.ylabel("WCSS")
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(k_values, silhouette_scores, marker='o', color='orange')
plt.title("Coeficiente de Silhouette")
plt.xlabel("Número de clusters (k)")
plt.ylabel("Silhouette")
plt.grid(True)

plt.tight_layout()
plt.show()
```
